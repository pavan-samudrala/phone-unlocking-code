```python
import time
correct_password='pavan123'
max_attempts=3
while True: #creates an infinite loop â€” a loop that keeps running forever unless we explicitly stop it.
    attempts=0
    while attempts<max_attempts:
        password=input("enter password: ")
        if password==correct_password:
            print("unlocked successfully")
            break
        else:
            attempts+=1
            print(f"incorrect password. Attempts left: {max_attempts - attempts}")
    if password==correct_password:
        break
    print("too many incorrect attempts")
    print("phone locked for 30 seconds")
    time.sleep(30)
    print("time is over")
```
